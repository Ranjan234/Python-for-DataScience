# ASSIGNMENT: PREPARE ROWS FOR MODELING
 1. Determine the row granularity needed
 2. Create a column called June Purchases that sums
all purchases in June
 3. Create a column called Total Spend that sums the
prices of the books purchased in April & May
 4. Combine the June Purchases and Total Spend
columns into a single DataFrame for modeling

```python
# What should one row of data look like for the model?
# Customer purchased in june or not - sumary stats about april and may
```

```python
df.head(5)
```

```python
june.head()
```

```python
# Create a column called 'june purchases' that sums of all purchases in june
june_purchases = (june.groupby('Customer ID')['Book']
      .count()
      .rename('June Purchases')
      .to_frame()
      .reset_index()
)

june_purchases.head()
```

```python
df.head()
```

```python
df_april_may = df[df['Purchase Date'].dt.month < 6]
df_april_may.tail()
```

```python
# Create a column called 'Total Spend' that sums of prices of all books that were purchased in april and may
total_spend = (df_april_may.groupby('Customer ID')['Price']
             .sum()
             .rename('Total Spend')
             .to_frame()
             .reset_index())
total_spend.head()
```

```python
# combine the 'June Purchases' and 'Total Spend' column into a single DataFrame for modeling
total_spend.merge(june_purchases, how ='left').fillna(0)
```

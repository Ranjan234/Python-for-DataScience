```python
import pandas as pd
```

```python
orders = pd.read_csv('/content/orders.csv', parse_dates=["order_date"] )
promotions = pd.read_csv('/content/promotions.csv', parse_dates=["start_date","end_date"])
```

```python
pd.merge_asof(
    orders.sort_values("order_date"),
    promotions.sort_values("start_date"),
    left_on="order_date",
    right_on="start_date",
    direction="backward"
).query("order_date <= end_date")
```

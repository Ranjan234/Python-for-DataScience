```python
import statsmodels.api as sm
from sklearn.linear_model import LinearRegression
```

```python
diamonds = pd.read_csv('/content/Diamonds Prices2022.csv')
```

```python
diamonds.head()
```

```python
x= diamonds[['carat']]
y= diamonds['price']
```

```python
x = sm.add_constant(diamonds['carat'])
y = diamonds['price']
```

```python
model = sm.OLS(y, x).fit()

model.summary()
```

```python
model.predict([1, 2])
```

```python
new_diamonds = pd.DataFrame({'carat': [0, .1, .3, .5, 1, 2, 3, 5]})
```

```python
model.predict(sm.add_constant(new_diamonds))
```

```python
model = LinearRegression(fit_intercept=False).fit(x, y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
model.score(x, y)
```

• R-squared values are bounded between 0 and 1 on training data

# HYPOTHESIS TEST
- Regression models include several hypothesis tests, including the F-test, that
indicates whether our model is significantly better at predicting our target than
using the mean of the target as the mode

- Steps for the hypothesis test:

1) State the null and alternative hypotheses

2) Set a significance level (α)

3) Calculate the test statistic and p-value

4) Draw a conclusion from the test

a) If p≤α, reject the null hypothesis (you’re confident the model isn’t useless)

b) If p>α, don’t reject it (the model is probably useless, and needs more training)

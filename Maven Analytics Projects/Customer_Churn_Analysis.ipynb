# Task 2: Churn Count Visualization

```python
# Use a Column chart to  display the count of customers in the churn column.
# Visualize the count of churned vs. non-churned customers.
plt.figure(figsize=(8,6))
sns.countplot(data=Customer, x='Churn')
plt.title('Count of Churned vs. Non -Churned Customer')
plt.show()
```

# Task 3: Monthly Charges Distribution


```python
# Create a histogram for the monthly charges column.
# Use appropriate bin sizes(e.g intervals of 10).
plt.figure(figsize=(8,6))
sns.histplot(data= Customer, x = 'MonthlyCharges', bins = 10)
plt.title('Monthly Charges Distribution')
plt.show()
```

# Task 4: Churn Rates by Gender

```python
# Compare churn rates across gender
# Create a Clusterd Bar chart showing churned and non- churned counts each gender.
churn_gender = Customer.groupby(['gender','Churn']).size().unstack()
churn_gender.plot(kind ='bar', stacked= True)
plt.title('Churn Rates By gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

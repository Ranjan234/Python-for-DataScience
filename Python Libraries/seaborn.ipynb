# Data Visualization with Seaborn
- Seaborn is a Python visualization library based on Matplotlib that provides a high-level interface for drawing attractive and informative statistical graphics. Seaborn helps in creating complex visualizations with just a few lines of code. In this lessons, we will cover the basics of seaborn, including creating various types of plots and customizing them.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
## Basic ploting with seaborn
sales = pd.read_csv('D:\Master Data Analytics BootCamp\Python_Libraries\data.csv')
```

```python
sales
```

```python
# create a scatterplot
sns.scatterplot(x = 'Sales',y = 'Product', data= sales)
plt.title("Scatterplot of total Sales vs product")
plt.show()
```

```python
## line plot
sns.lineplot(x = 'Sales',y = 'Region', data= sales)
plt.title("Lineplot of total Sales vs region")
plt.show()
```

```python
# ctaegorical plot
sns.barplot(x = 'Category',y = 'Value', data= sales)
plt.title("Categorical plot of total category vs value")
plt.show()
```

```python
## Box plot
sns.boxplot(x = 'Product',y = 'Sales', data= sales)
plt.title("Boxplot of total Sales vs product")
plt.show()
```

```python
# Vialin plot
sns.violinplot(x = 'Product',y = 'Sales', data= sales)
plt.title("Violin plot of total Sales vs product")
plt.show()
```

```python
## histogram
sns.histplot(sales['Sales'], kde=True)
plt.title("Histogram of total Sales")
plt.show()
```

```python
### KDE plot
sns.kdeplot(sales['Sales'], shade=True)
plt.title("KDE plot of total Sales")
plt.show()
```

```python
## Pairplot
sns.pairplot(sales)
plt.title("Pairplot of sales data")
plt.show()
```

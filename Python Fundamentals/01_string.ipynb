# Strings
- String is a data type in python that stores a sequence of characters - letters, numbers, or symbols- enclosed in single(''), double(""), triple(""" """) quotes.
- Note : Strings are immutable, meaning once created, their content cannot be changed directly.

```python
# Example

str1 = 'hello'
str2 = "world"
str3 = '''This is a
multi-line
string.'''  

print(str1)
print(str2)
print(str3)
```

```python
# string concatenation
greeting = str1 + ' ' + str2 + '!'
print(greeting)
```

```python
# length of string
length = len(greeting)
print(f'Length of greeting: {length}')
```

- Indexing :
  - Each character in a string has a position(index) starting from 0.
- Slicing :
  - Slicing lets you access a part of a string.
   - Str[strat : end]   

```python
str = "Soumya ranjan sahoo"
print(str[0])          # First character
print(str[5])          # Sixth character
print(str[-1])         # Last character
print(str[0:6])       # First six characters
print(str[7:])        # From eighth character to end
print(str[:6])        # From start to sixth character
print(str[-6:])       # Last six characters 
print(str[:])         # Entire string
```

```python
print(str.upper())      # Convert to uppercase
print(str.lower())      # Convert to lowercase
print(str.replace("sahoo", "Sahoo"))  # Replace substring
print(str.split(" "))   # Split string into list of words
print(str.find("ranjan"))  # Find substring index
print(str.count("a"))   # Count occurrences of substring
print(str.strip())     # Remove leading/trailing whitespace
print(str.startswith("Soumya"))  # Check start substring
print(str.endswith("sahoo"))    # Check end substring
print(str.isalpha())    # Check if all characters are alphabetic
print(str.isdigit())    # Check if all characters are digits]
print(str.isalnum())    # Check if all characters are alphanumeric
print(str.index("ranjan"))  # Get index of substring
print(str.capitalize())  # Capitalize first character
print(str.title())       # Title case

```

```python
print("Hello world!")
print("Hello \n world!")
print("Hello \t world!")
print("Hello \\ world!")
print("Hello \"world\"!")
```

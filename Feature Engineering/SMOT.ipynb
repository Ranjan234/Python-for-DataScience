## SMOTE
   ### Synthetic Minority Oversampling Techniques.
   - SMOTE is a technique used in machine learning to address imbalanced datasets where the minority class has signifiacantly fewer instances than the majority class. SMOTE instances generate synthetic instances of the minority class by interpolating between existing instances.

```python
from sklearn.datasets import make_classification
```

```python
x, y=make_classification(n_samples=1000,n_redundant=0, n_features =2, n_clusters_per_class=1, weights = [0.90],random_state =12 )
```

```python
import pandas as pd
df1 = pd.DataFrame(x, columns = ['f1','f2'])
df2 = pd.DataFrame(y, columns =['target'])
final_df = pd.concat([df1,df2],axis=1)
```

```python
final_df.head()
```

```python
# target count
final_df['target'].value_counts()
```

```python
import matplotlib.pyplot as plt
plt.scatter(final_df['f1'],final_df['f2'],c=final_df['target'])
```

```python
from imblearn.over_sampling import SMOTE
```

```python
# transform the dataset
oversampling = SMOTE()
x, y = oversampling.fit_resample(final_df[['f1','f2']], final_df['target'])
```

### Handling Outliers With Python
  # 5 number summary and Box plot

```python
import numpy as np
```

```python
# minimum, maximum, median, Q1,Q3, IQR
lst_marks= [45,32,56,75,89,54,32,89,90,87,67,54,45,98,99,67,74]
minimum,Q1,median,Q3,maximum=np.quantile(lst_marks,[0,0.25,0.50,0.75,1.0])
```

```python
minimum,Q1,median,Q3,maximum
```

```python
IQR = Q3-Q1
```

```python
print(IQR)
```

```python
lower_fence= Q1- 1.5*(IQR)
higher_fence= Q3+ 1.5*(IQR)
```

```python
print(lower_fence)
print(higher_fence)
```

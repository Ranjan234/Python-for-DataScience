- Example : A student, car, Bank Account, each can be represented as an object- in python.

```python
class Student:
    school_name = "ABC School"

    def __init__(self, name, age, grade):
        self.name = name
        self.age = age
        self.grade = grade

student1 = Student("Alice", 14, "8th")
print("Student Name:", student1.school_name)
```

```python
# Static method

# create class student that takes  3 marks and has a method average().
class Student:

    def __init__(self, name, listofmarks):
        self.name = name
        self.listofmarks = listofmarks

    #static method
    def average(self):
        print("The average marks are:", sum(self.listofmarks)/len(self.listofmarks))

student1 = Student("Soumya", [85, 90, 95])     
student1.average()   
print("Student Name:", student1.name)
print("Marks:", student1.listofmarks)  
```

# OOPS:
 - Object -oreinted programming (OOPs) is a structure programs using real-wolrd  concept like objects, classes, attributes,and behaviors. Instead of writting everything in one place, we organize code into objects.

```python
# class creation
class Student:
      name = "Soumya ranjan"  # attribute
      roll_no = 7
      age = 25
      sub = "math"
      course = "BCA"

# creating object
s1= Student()
print(s1.name)
print(s1.age)
print(s1.roll_no)
```

 - Encapsulation :
   - Wrraping data + methods inside a single unit (class). Data is protected using private variables. 

```python
# example of encapsualtion in a banking system and print account balanceclass Account:
class Bank:
    def __init__(self):
        self.accounts = {}

    def create_account(self, account_id, initial_balance=0):
        if account_id not in self.accounts:
            self.accounts[account_id] = Account(account_id, initial_balance)
            return True
        return False

    def deposit(self, account_id, amount):
        if account_id in self.accounts:
            return self.accounts[account_id].deposit(amount)
        return False

    def withdraw(self, account_id, amount):
        if account_id in self.accounts:
            return self.accounts[account_id].withdraw(amount)
        return False

    def get_balance(self, account_id):
        if account_id in self.accounts:
            return self.accounts[account_id].get_balance()
        return None



# Example usage
bank = Bank()
bank.create_account("12345", 1000)
bank.deposit("12345", 500)
bank.withdraw("12345", 200)

print("Account Balance:", bank.get_balance("12345"))  # Output: Account Balance: 1300


```

- Inheritance
 - When one class (child) gets the properties and methods of another class(parent). It avoids repeating code.
 

```python
# Parent class
class vehicle:
      def start(self):
          print("Vehicle started")

# child class1 
class car(vehicle):
    def drive(self):
        print("Car is driving")

# child class2
class bike(vehicle):
    def ride(self):
        print("Bike is riding")

# child class3
class truck(vehicle):
    def haul(self):
        print("Truck is hauling goods")

# Example usage
my_car = car()
my_car.start()  # Inherited method from vehicle
my_car.drive()

my_bike = bike()
my_bike.start()  # Inherited method from vehicle
my_bike.ride()

my_truck = truck()
my_truck.start()  # Inherited method from vehicle
my_truck.haul()
```

- Polymerphism
  - Same function name, but different behaviour in different classes.
  - One function name -many different behaviours depending on the object calling it.
  - "Same name, different action"
   - Same method name 
    - different output
    - depends on the objects
    

```python
# Example
class Cat:
    def sound(self):
        print("Cat says :Meow!")

class Dog:
    def sound(self):
        print("Dog says :Woof!")
 

class Cow:
    def sound(self):
        print("Cow says :Moo!")


#Polymerphism example
animals = [Cat(), Dog(), Cow()]               

for animal in animals:
    animal.sound()

```

```python
# Example 2- Shapes (Same method, diffeerent implementation)
class Circle:
    def area(self, radius):
        return 3.14 * radius * radius

class Square:
    def area(self, side):
        return side * side

class Rectangle:
    def area(self, length, width):
        return length * width

# Example usage
shapes = [Circle(), Square(), Rectangle()]
for shape in shapes:
    if isinstance(shape, Circle):
        print("Circle area:", shape.area(5))
    elif isinstance(shape, Square):
        print("Square area:", shape.area(4))
    elif isinstance(shape, Rectangle):
        print("Rectangle area:", shape.area(4, 6))
```
